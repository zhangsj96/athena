```python
import numpy as np
def lnPsat(T): 
  return -56.82 - 18946/T + 9.3974*np.log(T)
def Psat(T): 
  return np.exp(lnPsat(T))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(np.logspace(2, 4, 100), Psat(np.logspace(2, 4, 100)), label='Psat')
plt.loglog()
```

```python
def rho_sat(T):
  kb_over_mu_mH = 2749927.464896386;
  return 1./kb_over_mu_mH * Psat(T)/T;
```

```python
plt.plot(np.logspace(2, 4, 100), rho_sat(np.logspace(2, 4, 100)), label='Psat')
plt.loglog()
```
